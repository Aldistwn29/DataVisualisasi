# Menambahkan label sumbu x dan sumbu y

Dalam istilah awam, Label X dan Label Y adalah sebutan yg diberikan masing-masing untuk Sumbu X dan Sumbu Y. ini dapat ditambahkan ke grafik dengan menggunakan motode xlabel() dan ylabel()

syntax:
matplotlib.pyplot xlabel(xlabel, fontdict=None, labelpad=None, **kwargs)
matplotlib.pyplot ylabel(ylabel, fontdict=None, labelpad=None, **kwargs)

```python
import matplotlib.pyplot as plt
# initializinf the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize = 20, color="green")

# Adding label on the x-axis
plt.xlabel("Sumbu X")

# Adding label on the y-axis
plt.ylabel("Sumbu Y")

plt.show()
```

# Mentapkan Batas dan label Centang

Kita mungkin telah melihat bahwa Matplotlib secara otomatis menetapkan nilai dan penanda (titik) dari sumbu X dan Y. namun, dimungkinkan untuk mengatur batas dan penanda secara manual xlim() dan ylim() digunakan untuk mengatur batas sumbu X dan sumbu Y masing-masing. Demikian pula, fungsi xticks() dan ytiks() di gunakan untuk menyetel label centang.

```python
import matplotlib.pyplot as plt

# initializinf the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize = 20, color="green")

# Adding label on the x-axis
plt.xlabel("Sumbu X")

# Adding label on the y-axis
plt.ylabel("Sumbu Y")

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "there", "four"])

plt.show()
```

# Menambahkan legends

legends adalah  area yg menggambarkan elemen grafik. secara sederhana, ini mencerminkan data yg di tampilkan dalam sumbu Y. Biasanya muncul sebagai kotak yg berisi sampel kecil dari setiap warna pada grafik dan deskripsi kecil tentang arti data ini.
Atribut bbox_to_anchor(x,y) dari fungsi legend() digunakan untuk menentukan koordinat legend, dan atribut ncol mewakili jumlah kolom yg dimiliki legend. Nilai default adalah 1.

syantax:
matplotlib.pyplot.legend(["name1","name2"}, bbox_to_ancher=(x,y),ncol=1)

```python
import matplotlib.pyplot as plt

# initializinf the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize = 20, color="green")

# Adding label on the x-axis
plt.xlabel("Sumbu X")

# Adding label on the y-axis
plt.ylabel("Sumbu Y")

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "there", "four"])

# Adding legends
plt.legend(["GFG"])

plt.show()
```

# Figure Class

class matplotlib.figure.Figure(figsize=None, dpl=None, facecolor=None, edgcolor=None,linewidth=0.0,fremeon=None,subplotpars=None,tight_layout=None,constrained_layout=None)

```python
  # Python progras to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initiakizing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating a new figure with width = 7 inches
# and height = 4 inches with face color as
# green, edgecolor as red and the line width
# of the edge as 7
fig = plt.figure(figsize=(7, 4), facecolor='g',
                 edgecolor='orange', linewidth=10)

# Creating a new axes for the figure
ax = fig.add_axes((1, 1, 1, 1))

# Adding the data to be plotted
ax.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize=24, color="black")

# Adding label on the y-axis
plt.ylabel("Sumbu Y")

# Adding label on the x-label
plt.xlabel("Sumbu X")

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "thre", "four"])

# Addin lagends
plt.legend(["GFG"])

plt.show()
```

# Axes Class

class Axes adalah unit paling dasar dan fleksibel untuk membuat sub-plot.Gambar tertentu mungkin berisi banyak sumbu. tetapi sumbu tertentu hanya dapat ada dalam gambar. Fungsi axes() membuat objek axes.

```python
  # Python progras to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating a new figure with width = 6 inches
# and height = 4 inches
fig = plt.figure(figsize=(6, 4))

# Creating a new axes for the figure
ax = fig.add_axes((1, 1, 1, 1))

# Plotting 1st dataset to the figure
ax1 = ax.plot(x, y)

# Plotting 1st dataset to the figure
ax2 = ax.plot(y, x)

# Setting Tittle
ax.set_title("Grafik Linear")

# Setting Title
ax.set_xlabel("Sumbu X")
ax.set_ylabel("Sumbu Y")

# Addin lagends
ax.legend(labels= ("garis 1", "garis 2"))

plt.show()
```

# Multiple Plot(Plot lebih dari 1)

kita telah mempelajari tentang komponen dasar graf yg dapat di tambahkan sehingga dapat menyimpan informasi lebih lanjut. salah satu caranya adalah dengan memanggil fungsi plot berulang kali dengan kumpulan nilai yg berbeda seperti yg di tunjukkan pada contoh di atas. Sekarang mari kita lihat cara memplot banyak grafik menggunakan berbagai fungsi dan juga cara memplot subplot.

example 1

using add_axes()

```python
# Python progras to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating a new figure with width = 6 inches
# and hight = 4 inches
fig = plt.figure(figsize=(6, 4))

# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Creating sconde axes for the figure
ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])

# Adding the data to be plotted
ax1.plot(x, y)
ax2.plot(y, x)

plt.show()
```

```python
import matplotlib.pyplot as plt

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating figure object
plt.figure()

# adding first subplot
plt.subplot(121)
plt.plot(x, y)

# Adding seconds subplots
plt.subplot(122)
plt.plot(y, x)

plt.show()
```
